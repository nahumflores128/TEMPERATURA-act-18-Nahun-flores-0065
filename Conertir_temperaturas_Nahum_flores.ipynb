## 1. Fundamentos Teóricos
En el estudio de la termodinámica, las escalas Celsius ($C$) y Fahrenheit ($F$) se relacionan de forma lineal.

**Fórmulas de Conversión:**
* **Celsius a Fahrenheit:** $F = (C \times \frac{9}{5}) + 32$
  
* **Fahrenheit a Celsius:** $C = \frac{(F - 32) \times 5}{9}$

---

## 2. Ejemplos Prácticos

### Ejemplo A: Temperatura corporal (37°C) a °F
1. **Sustitución:** $F = (37 \times 1.8) + 32$
2. **Operación:** $66.6 + 32 = 98.6$
3. **Resultado:** 98.6°F

### Ejemplo B: 100°F a °C
1. **Sustitución:** $C = (100 - 32) \times 5/9$
2. **Operación:** $68 \times 0.555... = 37.77$
3. **Resultado:** 37.77°C

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("\nIngrese los grados Celsius: "))
    f = (c * 9/5) + 32
    print(f"--> {c}°C equivalen a {f:.2f}°F")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("\nIngrese los grados Fahrenheit: "))
    c = (f - 32) * 5/9
    print(f"--> {f}°F equivalen a {c:.2f}°C")

def menu():
    """Función principal que controla el flujo del programa"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Nahum flores NL = 18")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
